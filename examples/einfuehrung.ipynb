## Unfallschwere-Balkendiagramm

Visualisierung der Verunfallten nach Schweregrad mit der FaSi-eigenen Farbpalette.

```python
palette = get_theme_palette("unfallschwere")
jahre = [2019, 2020, 2021, 2022, 2023]
leicht = [1823, 1654, 1712, 1798, 1845]
schwer = [342, 298, 315, 328, 301]
getoetet = [28, 22, 24, 19, 21]

fig, ax = plt.subplots(figsize=(10, 5))
x = range(len(jahre))
width = 0.25
ax.bar([i - width for i in x], leicht, width, label='Leichtverletzte', color=palette['leichtverletzte'])
ax.bar(x, schwer, width, label='Schwerverletzte', color=palette['schwerverletzte'])
ax.bar([i + width for i in x], getoetet, width, label='Getötete', color=palette['getötete'])
ax.set_xticks(list(x))
ax.set_xticklabels([str(j) for j in jahre])
ax.legend()
ax.set_title('Verunfallte nach Schwere – Kanton Zürich (Fiktive Daten)')
ax.set_ylabel('Anzahl')
plt.tight_layout()
plt.savefig('beispiel_unfallschwere.png', dpi=100, bbox_inches='tight')
plt.show()
print("Quelle: Fiktive Musterdaten | FaSi ZH Viz")
```

## Kontrast-Check (WCAG 2.1)

Überprüfung aller Infografik-Farben auf WCAG-konformen Kontrast (≥ 3:1 auf Weiss).

```python
from fasi_zh_viz import validate_palette_against_background, load_tokens
tokens = load_tokens()
palette_colors = list(tokens["colors"]["infographics_palette"].values())
result = validate_palette_against_background(palette_colors, "#FFFFFF")
print("Alle Farben kontrastkonform:", result["ok"])
for entry in result.get("results", []):
    print(f"  {entry['color']}: {entry['ratio']:.2f}:1 – {'OK' if entry['ok'] else 'FAIL'}")
```
